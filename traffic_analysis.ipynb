# Traffic Collision Data Analysis
This notebook demonstrates basic analysis of road accident data using Python.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('accident_data_sample.csv')
df.head()
```

```python
# Plot accident severity count
severity_counts = df['Severity'].value_counts()
severity_counts.plot(kind='bar', title='Accident Severity Count', color='teal')
plt.xlabel('Severity')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()
```
